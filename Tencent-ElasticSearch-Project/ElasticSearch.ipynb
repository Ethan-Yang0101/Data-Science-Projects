```python
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch(HOST="http://localhost",PORT=9200)
es = Elasticsearch()
```

```python
es.indices.create(index="first_index",ignore=400)
```

```python
es.indices.exists(index="first_index")
```

```python
doc1 = {"city":"New Delhi","country":"India"}
doc2 = {"city":"London","country":"England"}
doc3 = {"city":"Los Angeles","country":"USA"}
```

```python
es.index(index="cities",doc_type="places",id=1,body=doc1)
```

```python
es.index(index="cities",doc_type="places",id=2,body=doc2)
```

```python
es.index(index="cities",doc_type="places",id=3,body=doc3)
```

```python
res = es.get(index="cities",doc_type="places",id=2)
res["_source"]
```

```python
doc1 = {"sentence":"Today is a sunny day"}
doc2 = {"sentence":"Today is a bright-sunny day"}
```

```python
es.index(index="english",doc_type="sentences",id=1,body=doc1)
```

```python
es.index(index="english",doc_type="sentences",id=2,body=doc2)
```

```python
res = es.search(index="english",body={"from":0,"size":0,"query":{"match":{"sentence":"SUNNY"}}})
res
```

```python
res = es.search(index="english",body={"from":0,"size":2,"query":{"match":{"sentence":"SUNNY"}}})
res["hits"]["hits"]
```

```python
res = es.search(index="english",body={"from":0,"size":2,"query":{"match_phrase":{"sentence":"SUNNY"}}})
res["hits"]["hits"]
```

```python
res = es.search(index="english",body={"from":0,"size":1,"query":{"bool":{"must_not":{"match":{"sentence":"bright"}}}}})
res["hits"]["hits"]
```

```python
res = es.search(index="english",body={"from":0,"size":2,"query":{"regexp":{"sentence":".*"}}})
res["hits"]["hits"]
```
