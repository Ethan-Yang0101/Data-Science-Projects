```
def get_title_and_label(filename):
  data = open(filename).read()
  contents = []
  for line in data.split("\n"):
    if line != "":
      content = line.split("\t")
      contents.append(content)
  dataset = [[content[4],content[7]] for content in contents]
  return dataset
```

```
dataset = get_title_and_label("20200913")
dataset[:5]
```

```
import pandas as pd
import numpy as np

title = ["新闻标题","一级类目"]
Data = pd.DataFrame(data=dataset, columns=title)
Data.head()
```

```
Data.info()
```

```
Data.dropna(inplace=True)
Data.drop_duplicates(inplace=True)
Data = Data[Data["一级类目"] != "unknow"]
top_10_label = pd.Series(Data["一级类目"]).value_counts()[:10].index.tolist()
top_10_Data = Data.loc[Data["一级类目"].isin(top_10_label),:]
top_10_Data.reset_index(drop=True, inplace=True)
```
