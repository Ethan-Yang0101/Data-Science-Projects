```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 对所有数据报告和矩阵做数据处理

```python
process_file_name = [
    'results/all_prf_cnn',
    'results/all_prf_dpcnn',
    'results/all_prf_drnn',
    'results/all_prf_fast',
    'results/all_prf_rcnn',
    'results/all_prf_rnn',
    'results/all_prf_trans'
]

process_confusion_name = [
    "results/all_confusion_matrix_cnn",
    "results/all_confusion_matrix_dpcnn",
    "results/all_confusion_matrix_drnn",
    "results/all_confusion_matrix_fast",
    "results/all_confusion_matrix_rcnn",
    "results/all_confusion_matrix_rnn",
    "results/all_confusion_matrix_trans"
]

model_name = ['CNN','DPCNN','DRNN','FastText','RCNN','RNN','Transformer']

model_result_dict = {}

for i, file_name in enumerate(process_file_name):
    data1 = open(file_name).read()
    data2 = open(process_confusion_name[i]).read()
    contents = []
    
    for line in data1.split("\n"):
        if line != "":
            content = line.replace(": ",", ").split(", ")
            contents.append(content)

    table1_title = ['Model','Class','Precision','Recall','F-Score']
    content = contents[2:]
    for cont in content:
        cont.insert(0,model_name[i])
    table1_data = pd.DataFrame(data=content)
    table1_data = table1_data.loc[:,[0,1,3,5,7]]
    table1_data.columns = table1_title
    
    line1 = contents[0][:7]
    line2 = contents[1][:7]
    line1.insert(0,model_name[i])
    line2.insert(0,model_name[i])
    table2_title = ['Model','Evaluation','Precision','Recall','F-Score']
    table2_data = pd.DataFrame(data=[line1,line2])
    table2_data = table2_data.loc[:,[0,1,3,5,7]]
    table2_data.columns = table2_title
    
    numbers = [int(s) for s in data2.replace("\n","").split('\t') if s.isdigit()]
    matrix = np.array(numbers).reshape(10,10)
    column_name = ["体","健","娱","房","政","教","社","科","财","车"]
    
    model_result_dict[model_name[i]] = (table2_data, table1_data, matrix, column_name)
```

### 生成整体模型评估数据

```python
model_perform = pd.concat([model_result_dict[name][0] for name in model_name])
```

```python
model_perform[model_perform["Evaluation"]=="macro_average"].sort_values(by="F-Score",ascending=False)
```
