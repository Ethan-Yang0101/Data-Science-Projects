```
def get_data(filename):
  data = open(filename).read()
  dataset = []
  for id, line in enumerate(data.split("\n")):
    if line != "":
      content = line.split("\t")
      content.insert(0, id)
      dataset.append(content)
  return dataset
```

```
dataset = get_data("raw_data.txt")
dataset[:5]
```

```
import pandas as pd
import numpy as np

title = ["搜索ID", "语义1", "语义2", "匹配与否"]
Data = pd.DataFrame(data=dataset, columns=title)
Data.head()
```

```
Data.info()
```

```
import re

def clean_data(sentence):
    clean_punc_text = re.sub(r'[^\w\s]','',sentence).replace(" ","").strip()
    clean_punc_num_text = re.sub(r'[0-9]+','',clean_punc_text)
    clean_punc_num_eng_text = re.sub(r'[a-zA-Z]+', '', clean_punc_num_text) 
    return clean_punc_num_eng_text

Data["语义1"] = Data["语义1"].apply(clean_data)
Data["语义2"] = Data["语义2"].apply(clean_data)
Data.head()
```

```
Data.dropna(inplace=True)
Data.drop_duplicates(inplace=True)
Data['语义1'] = Data['语义1'].replace('', np.nan)
Data["语义2"] = Data["语义2"].replace('', np.nan)
Data.dropna(subset=['语义1'],inplace=True)
Data.dropna(subset=["语义2"],inplace=True)
Data.reset_index(drop=True, inplace=True)
```
